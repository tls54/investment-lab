# Test yfinance api

```python
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
# Define the ticker symbol for Apple (AAPL)
ticker_symbol = "AAPL"

# Download Apple stock data from Yahoo Finance
try:
    data = yf.download(ticker_symbol, start="2023-01-01", end="2024-01-01")  # Adjust dates as needed
except Exception as e:
    print(f"Error downloading data: {e}")
    exit()

# Check if data was successfully downloaded
if data.empty:
    print("No data downloaded. Check the ticker symbol and date range.")
    exit()

# Plot the closing prices
plt.figure(figsize=(12, 6))  # Adjust figure size for better visualization
plt.plot(data['Close'], label=f'{ticker_symbol} Closing Price')
plt.title(f'{ticker_symbol} Stock Price')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)  # Add grid for better readability
plt.show()

# Optional:  Print some basic statistics
print("\nBasic Statistics:")
print(data.describe())
```
